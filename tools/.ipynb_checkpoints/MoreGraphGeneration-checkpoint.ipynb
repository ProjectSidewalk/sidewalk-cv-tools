```python
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
import csv
import os
from datetime import date
from scipy import stats

today = str(date.today())
numberofboxes = 20
cover = 100.0/numberofboxes

%matplotlib inline

class Range: 

    def __init__(self, lower_bound, upper_bound):
        self.lower = lower_bound
        self.upper  = upper_bound
        self.labels = ["NoCurbRamp", "Null", "Obstacle", "CurbRamp", "SurfaceProblem"]
        features = len(self.labels)
        self.total = [0 for i in range(features)]
        self.accurate = [0 for i in range(features)]

    def inrange(self, value):
        return (value > self.lower and value <= self.upper)

    def addtodata(self, label, match):
        val = self.labels.index(label)
        self.total[val] += 1
        if(match):
            self.accurate[val] += 1
    
    def toString(self):
        return (str(self.total) + "\n" + str(self.accurate))
    
    def getvalfromlabel(self, label):
        con = self.labels.index(label)
        return (self.accurate[con], self.total[con])
    
    def getaverage(self):
        return (self.lower + self.upper)/2




def create_ranges():
    ranges = []
    current = 0.0
    while(current < 100.0):
        copy = Range(current, min(current + cover, 100.0))
        current += cover 
        ranges.append(copy)
    return ranges
```

```python
colors = ['blue','green','red','orange','yellow','purple']
labels = ["NoCurbRamp","Obstacle", "CurbRamp", "SurfaceProblem"]

lower = 1
upper = 6

minconfidence = 5
maxconfidence = 100
increment = 5

def make_scatter(x,y, count): 
    plt.xlim(0,100)
    plt.ylim(0, 60)
    plt.plot(x,y, color = colors[count], marker='o', markerfacecolor= colors[count])

def import_from_file(number, ranges):
    path = os.path.join("single", "summary" + str(number) + ".csv")
    with open(path, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        count = 0
        for row in csvreader:
            count += 1
            label = row[3]
            accurate = row[4] == label
            confidence = float(row[5])
            for val in ranges: 
                if val.inrange(confidence):
                    val.addtodata(label, accurate)
        print(str(number) + " has " + str(count))

def make_all_ranges():
    all_ranges = []
    for i in range(lower, upper + 1):
        ranges = create_ranges()
        import_from_file(i,ranges)
        all_ranges.append(ranges)
    return all_ranges 

def make_graph():
    all_ranges = make_all_ranges()
    for label in labels:
        legend = []
        for number in range(lower, upper + 1):
            ranges = all_ranges[number - lower]
            x = []
            y = []
            complete = 0
            for current_confidence in range(minconfidence,maxconfidence,increment):
                works = 0
                total = 0
                for val in ranges:
                    (num, com) = val.getvalfromlabel(label)
                    total += com 
                    if(val.getaverage() >= current_confidence):
                        works += num
                if(total != 0):
                    x.append(current_confidence)
                    percentage = round((100.0 * works)/total)
                    y.append(percentage)
                if(complete == 0):
                    complete += total
            if(len(x) != 0):
                make_scatter(x,y,number - lower)
                t = "Minimum " + str(number) + " validators (" + str(complete) + " labels)"
                legend.append(t)
        fig = plt.gcf()
        fig.set_size_inches(8,4,forward=True)
        plt.xlabel('CV confidence threshold')
        plt.ylabel('Perecentage of labels above threshold')
        plt.title(label)
        plt.legend(legend)
        name = "Model_" + label + ".png"
        path = os.path.join("viz",name)
        plt.savefig(path)
        plt.show()
        
make_graph() 
```
